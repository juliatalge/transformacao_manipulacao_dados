```python
import pandas as pd
```

```python
dados = pd.read_json('/content/dados_hospedagem.json')
dados.head()
```

```python
#normalização dos dados com json_normalize
dados = pd.json_normalize(dados['info_moveis'])
dados
```

```python
#listagem dos nomes das colunas
colunas = list(dados.columns)
colunas
```

```python
#usando o .explode() para desagrupar dados
dados = dados.explode(colunas[3:])
dados
```

```python
#resetar os índices para ordenação de dados - inplace = True substitui os indices no dataframe
dados.reset_index(inplace=True, drop=True)
dados.head()
```

```python
#verificar os tipos de dados
dados.info()
```

```python
#conversão de dados pelas colunas numéricas
import  numpy as np
```

## Tratando dados númericos - convertendo para tipo inteiro

```python
#tratando dados numéricos - convertendo para tipo inteiro
dados['max_hospedes'].astype(np.int64)
```

```python
#atribuição dos dados
dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64)
```

```python
dados.info()
```

```python
col_numericas = ['quantidade_banheiros','quantidade_quartos','quantidade_camas']
```

```python
dados[col_numericas] = dados[col_numericas].astype(np.int64)
```

```python
dados.info()
```

## Tratando dados númericos - convertendo para o tipo float

```python
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
```

```python
dados.info()
```

```python
#caracteres como cifrão e vírgula não podem ser convertidos para o tipo float (exemplo: coluna preco)
dados['preco']
```

```python
#o método .apply() com lambda, que aplica uma função específica a cada elemento, no caso, a remoção dos caracteres
#Dentro de .apply(), passamos uma função lambda que lerá cada elemento x, ou seja, os dados de cada linha. Feito isso, usaremos o método .replace() para substituir os caracteres de cifrão e vírgula por nada, ou seja, por uma string vazia. Depois, usaremos .strip() para remover qualquer caractere antes ou depois do elemento
dados['preco'].apply(lambda x: x.replace('$', '').replace(',','').strip())
```

```python
#atribuindo o comando a coluna preco
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$', '').replace(',','').strip())
```

```python
dados['preco'] = dados['preco'].astype(np.float64)
```

```python
dados.info()
```

```python
#conversao de duas colunas
#começar pela visualização
dados[['taxa_deposito','taxa_limpeza']]
```

```python
#Quando tratamos da conversão de um dataframe, no qual passamos por cada elemento, é mais viável utilizar o método .applymap()
dados[['taxa_deposito','taxa_limpeza']].applymap(lambda x: float(x.replace('$', '').replace(',','').strip()))
```

```python
#atribuir os valores as colunas em questão
dados[['taxa_deposito','taxa_limpeza']] = dados[['taxa_deposito','taxa_limpeza']].applymap(lambda x: x.replace('$', '').replace(',','').strip())
```

```python
dados[['taxa_deposito','taxa_limpeza']] = dados[['taxa_deposito','taxa_limpeza']].astype(np.float64)
```

```python
dados.info()
```

## Manipulação dos textos para tokenização

```python
dados['descricao_local']
```

```python
#fazer o processo de leitura dos elementos da coluna como strings
#para transformar as letras em minusculas usar o método lower()
dados['descricao_local'].str.lower()
```

```python
#fazer atribuição
dados['descricao_local'] = dados['descricao_local'].str.lower()
```

```python
dados.head()
```

### utilização do Regex pois é utilizado para fazer busca, validação e manipulação de dados porque, a partir dessa sequência padronizada de caracteres é possível definir os caracteres que queremos ou não selecionar.

```python
dados['descricao_local'][3169]
```

```python
#substituição dos caracteres especiais
#a expressão [^a-zA-Z0-9\-\'] diz selecione todos os caracteres, exceto as letras (maiúsculas ou minúsculas), números, hífen e apóstrofo
dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
```

```python
#atribuição para salvar no dataframe
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
```

## Removendo hifens fora de palavras compostas

### (?<!\w)-(?!\w)

*   (?<!\w): uma "lookbehind" negativa que verifica se há um caractere de palavra antes do hífen ou se há apenas um espaço em branco, resultando em "verdadeiro" se não houver nada;
*   (?!\w): uma lookahead negativa que verifica se há um caractere de palavra depois do hífen.



```python
dados['descricao_local'].str.replace('(?<!\w)-(?!\w)', '', regex=True)
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w)', '', regex=True)
```

## Criação de tokens

```python
#Para criar tokens, basta separar cada uma das palavras como elementos de uma lista. Para transformar uma string numa lista, usamos o método split().
dados['descricao_local'].str.split()
```

```python
dados['descricao_local'] = dados['descricao_local'].str.split()
dados.head()
```

```python
#tratamento dos dados
#no caso da coluna comodidades precisamos tirar as chaves e aspas
dados['comodidades'].str.replace('\{|}|\"','',regex=True)
```

```python
#atribuir a dados['comodidades']
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"','',regex=True)
```

```python
#criação da lista de tokens para a coluna comodidades
dados['comodidades'] = dados['comodidades'].str.split(',')
dados.head()
```

```python
#conjunto de meses agrupados e a quantidade de valores true
dt_data.groupby(dt_data['data'].dt.strftime('%Y-%m'))['vaga_disponivel'].sum()
```

```python
#armazenar os dados com subset
subset = dt_data.groupby(dt_data['data'].dt.strftime('%Y-%m'))['vaga_disponivel'].sum()
subset
```
